## **Phase 2: Intermediate (2–4 months) – Scaling & Reliability**

**Goal:** Learn to handle **more users, higher load, reliability**

**Skills / Concepts:**

* Load balancing (round-robin, least connections)
* Database replication & sharding
* Caching strategies (write-through, write-back)
* Queues / message brokers (Kafka, RabbitMQ)
* Monitoring & logging (Prometheus, Grafana, ELK stack)
* High-level trade-offs: consistency vs availability

**Projects:**

1. Chat App (1:1 messaging, store messages in DB, implement async delivery)
2. File Storage / Image Upload Service (with caching, CDNs, and background processing)

**Resources:**

* Gaurav Sen – YouTube intermediate
* Designing Data-Intensive Applications (book)



## **Phase 3: Advanced (4–6 months) – FAANG-Style Systems**

**Goal:** Build **complex, scalable, distributed systems**

**Skills / Concepts:**

* Microservices architecture
* Event-driven design & pub/sub
* Horizontal scaling for APIs & DB
* CAP theorem & trade-offs
* Rate limiting, throttling, circuit breakers
* Real-time streaming & analytics (Kafka/Spark)

**Projects:**

1. Social Media Feed (like Twitter) – cache, timeline generation, DB sharding
2. Video Streaming Service (like YouTube) – chunking, CDN, async encoding
3. ML Deployment Pipeline – train → version → deploy → monitor

**Resources:**
