# LinkedList
 **3. Notes on Complexity**

| Operation        | Singly Linked List |
| ---------------- | ------------------ |
| Append at end    | O(n)               |
| Prepend at start | O(1)               |
| Delete by value  | O(n)               |
| Search           | O(n)               |
| Traverse         | O(n)               |



````python
# Your LinkedList classes
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, val):
        new_node = Node(val)
        if not self.head:
            self.head = new_node
            return
        cur = self.head
        while cur.next:
            cur = cur.next
        cur.next = new_node

    def print_list(self):
        cur = self.head
        while cur:
            print(cur.val, end=" -> ")
            cur = cur.next
        print("None")

    # Extra: prepend (add at start)
    def prepend(self, val):
        new_node = Node(val)
        new_node.next = self.head
        self.head = new_node

    # Extra: delete by value
    def delete(self, val):
        cur = self.head
        if cur and cur.val == val:
            self.head = cur.next
            return
        prev = None
        while cur and cur.val != val:
            prev = cur
            cur = cur.next
        if cur:
            prev.next = cur.next

# === Usage ===
ll = LinkedList()
ll.append(10)
ll.append(20)
ll.append(30)
ll.print_list()   # 10 -> 20 -> 30 -> None

ll.prepend(5)
ll.print_list()   # 5 -> 10 -> 20 -> 30 -> None

ll.delete(20)
ll.print_list()   # 5 -> 10 -> 30 -> None
````

#  Trees
**4. Notes on Complexity**

| Operation          | Time Complexity                         |
| ------------------ | --------------------------------------- |
| Traverse all nodes | O(n)                                    |
| Search/Insert      | O(n) for general tree, O(log n) for BST |
| Space (recursive)  | O(h) → height of tree                   |

````python
# Your TreeNode class and traversals
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def inorder(node):
    if node:
        inorder(node.left)
        print(node.val, end=" ")
        inorder(node.right)

def preorder(node):
    if node:
        print(node.val, end=" ")
        preorder(node.left)
        preorder(node.right)

def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.val, end=" ")

# === Usage ===
# Create tree
#        1
#       / \
#      2   3
#     / \
#    4   5

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

print("Inorder Traversal:")
inorder(root)    # 4 2 5 1 3
print("\nPreorder Traversal:")
preorder(root)   # 1 2 4 5 3
print("\nPostorder Traversal:")
postorder(root)  # 4 5 2 3 1
````

````python
# Graph (Adjacency List)
graph = {
    'A': ['B','C'],
    'B': ['A','D'],
    'C': ['A','D'],
    'D': ['B','C']
}
````

````python
from collections import deque



# === BFS ===
def bfs(graph, start):
    visited = set()
    q = deque([start])
    while q:
        node = q.popleft()
        if node not in visited:
            print(node, end=" ")
            visited.add(node)
            q.extend([n for n in graph[node] if n not in visited])

# === DFS ===
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    if start not in visited:
        print(start, end=" ")
        visited.add(start)
        for n in graph[start]:
            dfs(graph, n, visited)

# === Usage ===
print("BFS starting from A:")
bfs(graph, 'A')  # A B C D
print("\nDFS starting from A:")
dfs(graph, 'A')  # A B D C (order may vary)
````

````python
# Each tuple: (parent, child)
edges = [
    (1,2),
    (1,3),
    (2,4),
    (2,5),
    (3,6)
]
````

````python
# Convert to Adjacency List
from collections import defaultdict

def to_adjacency_list(edges):
    adj_list = defaultdict(list)
    for u,v in edges:
        adj_list[u].append(v)
        # For undirected graph, also do: adj_list[v].append(u)
    return adj_list

adj_list = to_adjacency_list(edges)
print(dict(adj_list))
# Output: {1: [2,3], 2:[4,5], 3:[6]}
````

````python
def to_adjacency_matrix(edges):
    # Find max node value to size matrix
    nodes = set()
    for u,v in edges:
        nodes.add(u)
        nodes.add(v)
    n = max(nodes) + 1  # +1 for 0-index (optional)
    
    # Initialize matrix
    matrix = [[0]*n for _ in range(n)]
    
    # Fill edges
    for u,v in edges:
        matrix[u][v] = 1
        # For undirected: matrix[v][u] = 1
    return matrix

adj_matrix = to_adjacency_matrix(edges)
for row in adj_matrix[1:]:  # skipping index 0 if nodes start at 1
    print(row[1:])
````

````python
from collections import defaultdict

class Graph:
    def __init__(self, edges=[], directed=True):
        self.edges = edges
        self.directed = directed
        self.nodes = set()
        for u,v in edges:
            self.nodes.add(u)
            self.nodes.add(v)

    # Parent-Child dictionary
    def parent_child_dict(self):
        pc = defaultdict(list)
        for u,v in self.edges:
            pc[u].append(v)
        return dict(pc)

    # Adjacency List
    def adjacency_list(self):
        adj = defaultdict(list)
        for u,v in self.edges:
            adj[u].append(v)
            if not self.directed:
                adj[v].append(u)
        return dict(adj)

    # Adjacency Matrix
    def adjacency_matrix(self):
        n = max(self.nodes) + 1  # Assuming nodes are 1-indexed
        matrix = [[0]*n for _ in range(n)]
        for u,v in self.edges:
            matrix[u][v] = 1
            if not self.directed:
                matrix[v][u] = 1
        return matrix

# === Usage ===
edges = [(1,2),(1,3),(2,4),(2,5),(3,6)]
g = Graph(edges, directed=True)

print("Parent-Child Dict:")
print(g.parent_child_dict())

print("\nAdjacency List:")
print(g.adjacency_list())

print("\nAdjacency Matrix:")
matrix = g.adjacency_matrix()
for row in matrix[1:]:   # skipping 0 index if nodes start at 1
    print(row[1:])
````

````python
# Using dict (hashmap)  O(n)
def char_frequency(s): 
    freq = {}
    for c in s:
        if c in freq:
            freq[c] += 1
        else:
            freq[c] = 1
    return freq

s = "abracadabra"
print(char_frequency(s))
# Output: {'a': 5, 'b': 2, 'r': 2, 'c': 1, 'd': 1}

# Using Counter from collections
from collections import Counter
freq2 = Counter(s)
print(freq2)
# Output: Counter({'a': 5, 'b': 2, 'r': 2, 'c': 1, 'd': 1})
````

````python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, arr):
        for val in arr:
            new_node = Node(val)
            if not self.head:
                self.head = new_node
                cur = self.head
            else:
                cur.next = new_node
                cur = new_node

ll = LinkedList()
ll.append([1,2,3])
print(ll)
````
